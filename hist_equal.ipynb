```python
%matplotlib inline
from matplotlib import mlab
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
from scipy import ndimage
```

```python
im1 = Image.open('/media/yi/DATA/data-orig/robot/1/5/0.jpg').convert('RGB')
# im1 = Image.open('/home/yi/Downloads/VIPER/train/img/1/026/026_00001.jpg').convert('RGB')
im1 = np.array(im1)
img1 = im1 / 255.0
im2 = Image.open('/media/yi/DATA/data-orig/robot/1/5/1.jpg').convert('RGB')
# im2 = Image.open('/home/yi/Downloads/VIPER/train/img/1/026/026_00002.jpg').convert('RGB')
im2 = np.array(im2)
img2 = im2 / 255.0
plt.figure()
plt.imshow(img1)
plt.figure()
plt.imshow(img2)
```

```python
plt.figure()
plt.imshow(np.abs(img2 - img1))
```

```python
plt.figure()
plt.imshow(np.abs(img2[:, :, 0] - img1[:, :, 0]))
```

```python
img2 - img1
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1_0 = clahe.apply(im1[:, :, 0])
cl1_1 = clahe.apply(im1[:, :, 1])
cl1_2 = clahe.apply(im1[:, :, 2])
cl1 = np.dstack((cl1_0, cl1_1, cl1_2))
cl1 = cl1 / 255.0
cl2_0 = clahe.apply(im2[:, :, 0])
cl2_1 = clahe.apply(im2[:, :, 1])
cl2_2 = clahe.apply(im2[:, :, 2])
cl2 = np.dstack((cl2_0, cl2_1, cl2_2))
cl2 = cl2 / 255.0
plt.figure()
plt.imshow(cl1)
plt.figure()
plt.imshow(cl2)
```

```python
plt.figure()
plt.imshow(np.abs(cl2 - cl1))
```

```python
plt.figure()
plt.imshow(np.abs(cl2[:, :, 0] - cl1[:, :, 0]))
```

```python
cl2 - cl1
```

```python
plt.figure()
plt.imshow(np.abs(cl2[0:10, 0:10, 0] - cl1[0:10, 0:10, 0]))
```

```python
x1 = ndimage.median_filter(cl1, 5)
x2 = ndimage.median_filter(cl2, 5)
plt.figure()
plt.imshow(x1)
plt.figure()
plt.imshow(x2)
```

```python
plt.figure()
plt.imshow(np.abs(x2[:, :, 0] - x1[:, :, 0]))
```

```python
plt.figure()
plt.imshow(np.abs(x2[0:10, 0:10, 0] - x1[0:10, 0:10, 0]))
```
