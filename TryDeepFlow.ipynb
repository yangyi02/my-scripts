```python
flows = []
for i in xrange(10):
    flow = deepflow2(images[i], images[i+1])
    flows.append(flow)
```

```python
flow_mags = []
for i in xrange(10):
    flow_mag = numpy.sqrt(numpy.square(flows[i][:,:,0]) + numpy.square(flows[i][:,:,1]))
    flow_mags.append(flow_mag)
```

```python
for i in xrange(10):
    plt.imshow(flow_mags[i])
    plt.show()
```

```python
flow_grads = []
for i in xrange(10):
    flow_grad = numpy.gradient(flow_mags[i])
    flow_grad_mag = numpy.sqrt(numpy.square(flow_grad[0]) + numpy.square(flow_grad[1]))
    flow_grads.append(flow_grad_mag)
```

```python
for i in xrange(10):
    plt.imshow(flow_grads[i])
    plt.show()
```
